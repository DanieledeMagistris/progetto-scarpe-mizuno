```python
# trasformiamo le collone genere e modello in minuscolli
dfd['genere'] = dfd['genere'].str.lower()
dfd['modello'] = dfd['modello'].str.lower()
```

```python
#trasformiamo dentro la collona marca 'PUMA GOLF' in 'PUMA'
dfd['marca'] = dfd['marca'].str.replace('GOLF', '')
```

```python
# trasformiamo tutte le righe dalla collona 'sport' in minuscolo
df.loc[:,'sport'] = df['sport'].str.lower()
```

```python
# trasformiamo tutte le righe dalla collona 'sport' in minuscolo
dfd.loc[:,'sport'] = dfd['sport'].str.lower()
```

```python
df['sport'].value_counts()
```

```python
dfd['sport'].value_counts()
```

```python
#eliminiamo tutte le righe dove sport contiene 'sneakers basse'
df = df[~df['sport'].str.contains('sneakers basse', case=False, na=False)]
df['sport'].value_counts()
```

```python
#eliminiamo tutte le righe dove sport contiene 'sneakers basse'
dfd = dfd[~dfd['sport'].str.contains('sneakers basse', case=False, na=False)]
dfd['sport'].value_counts()
```

```python
# Salva il risultato in un nuovo file CSV
df.to_csv("Splittato_tutto.csv", index=False)
dfd.to_csv("Colori_aggregati.csv", index=False)

# Stampa una conferma
print("File unito e salvato come 'Scarpe_Sportive_Splittato_tutto.csv'")
```

```python
df = pd.read_csv("Splittato_tutto.csv", encoding='utf-8')
```

```python
df
```

```python
# Convertiamo 'nome_prodotto' in minuscolo
df['nome_prodotto'] = df['nome_prodotto'].str.lower()

# Verifichiamo se 'nome_prodotto' contiene 'unisex' e aggiorna 'genere'
df.loc[df['nome_prodotto'].str.contains('unisex'), 'genere'] = 'unisex'

# Visualizziamo il DataFrame
display(df)
```

```python
# Contiamo le righe che contengono "unisex" nella colonna "genere"
conteggio_unisex = len(df[df['genere'] == 'unisex'])

# Visualizziamo il conteggio
print(f"Ci sono {conteggio_unisex} righe con genere 'unisex'.")
```

```python
# Creamo un nuovo DataFrame senza le righe con "unisex" in "genere"
df_senza_unisex = df[df['genere'] != 'unisex']

# Visualizziamo il DataFrame risultante
print(display(df_senza_unisex))
```

```python
# Salva il risultato in un nuovo file CSV
df.to_csv("Splittato_tutto.csv", index=False)

# Stampa una conferma
print("File unito e salvato come 'Scarpe_Sportive_Splittato_tutto.csv'")
```
